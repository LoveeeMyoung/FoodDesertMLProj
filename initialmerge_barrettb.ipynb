```python
disease_df1 = disease_df1[disease_df1['year'] == 2017]
grocery_df1 = grocery_df1[grocery_df1['year'] == 2017]
grocery_df1['year'].unique()
disease_df1['year'].unique()
```

```python

def merging(disease_df1,grocery_df1):
    grocery_df1['tract_fips10'] = grocery_df1['tract_fips10'].astype(str)
    grocery_df1['tract_fips_last_five'] = grocery_df1['tract_fips10'].str[-5:]
    disease_df1['fips'] = disease_df1['fips'].astype(str).str.replace('.0', '')
    merged_df = pd.merge(grocery_df1, disease_df1, left_on='tract_fips_last_five', right_on='fips', how='inner')
    merged_df = merged_df.drop(columns=['tract_fips_last_five'])
    merged_df = merged_df.reset_index(drop=True)
    return merged_df
df = merging(disease_df1=disease_df1,grocery_df1=grocery_df1)
```
